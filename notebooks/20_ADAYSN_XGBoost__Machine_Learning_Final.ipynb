### <font color = '#6495ED'>**Introduction**</font>

#### *Objective:*
The objective of this project is to develop a predictive model to assist the New York Workers’ Compensation Board (WCB) in automating decision-making when new claims are received. The model will analyze various contributing factors, including individual characteristics, key dates related to the case, and specific details of each claim.
By leveraging data-driven insights, this project aims to identify significant correlations and patterns to build a reliable tool that accurately predicts the WCB’s final decision on the appropriate *Claim Injury Type* for a claim.
#### *Data:*
- Training Dataset: Contains data collected from the start of 2020 to the end of 2022.
- Test Dataset: Includes data with the same descriptive attributes as the training dataset, covering claims from January 2023 onward.
#### *Methodology:*
The project will follow these steps:
- Exploratory Data Analysis (EDA): Perform both univariate and bivariate analyses to understand the data distribution and relationships between variables.
- Multiclass Classification Benchmarking
- Model Optimization
#### *Success Metrics:*
Our primary metric will be F1 Score, as we are dealing with an imbalanced class problem.

#### <font color = '#6495ED'> **1.2. Data Analysis Helper Functions** </font>

```python
# define a function to analyse cardinality
def categorical_summary(df):
    # get categorical columns
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns

    # loop through each column and print its details
    for col in categorical_columns:
        print(f'#### {col}')
        print(f"Unique values count: {df[col].nunique()}")
        print(f"Unique values: {df[col].unique()}")
        print("\n")
```

```python
# define a function to analyse numerical column details
def numerical_summary(df):
    # get numerical columns
    numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

    # loop through each column and print its details
    for col in numerical_columns:
        print(f'#### {col}')
        print(f"Unique values count: {df[col].nunique()}")
        print(f"Minimum value: {df[col].min()}")
        print(f"Maximum value: {df[col].max()}")
        print(f"Mean: {df[col].mean()}")
        print(f"Median: {df[col].median()}")
        print(f"Standard Deviation: {df[col].std()}")
        print(f"Data type: {df[col].dtype}")
        print("\n")
```

```python
# create function to check percentage of missing values over time
def missing_perc(data):
    missing_values = round((data.isnull().sum()/data.shape[0])*100,2)
    print(missing_values)
```

```python
def categorical_summary(df):
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns

    # collect summary data for all categorical columns
    summary_data = []
    for col in categorical_columns:
        summary_data.append({
            'Column Name': col,
            'Unique Count': df[col].nunique(),
            'Unique Values': df[col].unique()[:5],
            'Data Type': df[col].dtype
        })
    # convert int into a dataframe
    # for better display
    summary_df = pd.DataFrame(summary_data)
    print("\nCategorical Summary:\n")
    print(summary_df.to_markdown(index=False))

categorical_summary(X_train)
```

Decision Trees for Bin Determination
Fit a decision tree for each numeric feature (Age at Injury, Average Weekly Wage) and visualize the tree to find the best splits for bins.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Function to find optimal bins and visualize tree
def find_bins_and_visualize(X, y, feature, max_depth=3):
    tree = DecisionTreeClassifier(max_depth=max_depth, random_state=42)
    tree.fit(X[[feature]], y)

    # Extract split points
    thresholds = tree.tree_.threshold[tree.tree_.feature == 0]
    thresholds = sorted(thresholds[thresholds != -2])  # Exclude invalid thresholds
    bins = [-np.inf] + thresholds + [np.inf]  # Add infinite bounds to cover all ranges

    # Visualize the tree
    plt.figure(figsize=(12, 6))
    plot_tree(tree, feature_names=[feature], filled=True, fontsize=10)
    plt.title(f"Decision Tree for {feature}")
    plt.show()

    return bins

# Determine and visualize bins for Age and Wage
age_bins = find_bins_and_visualize(X_train, y_train, 'Age at Injury')
wage_bins = find_bins_and_visualize(X_train, y_train, 'Average Weekly Wage')
IME4_bins = find_bins_and_visualize(X_train, y_train, 'IME-4 Count')

# Apply bins to datasets
X_train['Age Group'] = pd.cut(X_train['Age at Injury'], bins=age_bins)
X_train['Wage Group'] = pd.cut(X_train['Average Weekly Wage'], bins=wage_bins)
X_train['IME4 Group'] = pd.cut(X_train['IME-4 Count'], bins=wage_bins)

X_val['Age Group'] = pd.cut(X_val['Age at Injury'], bins=age_bins)
X_val['Wage Group'] = pd.cut(X_val['Average Weekly Wage'], bins=wage_bins)
X_val['IME4 Group'] = pd.cut(X_val['IME-4 Count'], bins=wage_bins)

X_test['Age Group'] = pd.cut(X_test['Age at Injury'], bins=age_bins)
X_test['Wage Group'] = pd.cut(X_test['Average Weekly Wage'], bins=wage_bins)
X_test['IME4 Group'] = pd.cut(X_test['IME-4 Count'], bins=wage_bins)
```

```python
# Mapping 'Industry Code' to new categories
industry_code_mapping = {
    'Education & Health': [61.0, 62.0],
    'Leisure & Hospitality': [71.0, 72.0],
    'Trade & Transportation': [42.0, 44.0, 45.0, 48.0],
    'Manufacturing & Construction': [23.0, 31.0, 32.0, 33.0],
    'Agriculture, Mining, Utilities': [11.0, 21.0, 22.0],
    'Information & Finance': [51.0, 52.0, 53.0],
    'Professional Services': [54.0, 55.0, 56.0],
    'Public & Other Services': [81.0, 92.0],
}

# Function to categorize 'Industry Code'
def categorize_industry_code(code):
    if pd.isna(code):  # explicitly handle missing values
        return 'Unknown'
    for category, codes in industry_code_mapping.items():
        if code in codes:
            return category
    return 'Other'  # for codes not included in the mapping

# Apply the function to train, test, and validation datasets
X_train['Industry Category'] = X_train['Industry Code'].apply(categorize_industry_code)
X_test['Industry Category'] = X_test['Industry Code'].apply(categorize_industry_code)
X_val['Industry Category'] = X_val['Industry Code'].apply(categorize_industry_code)
```

```python
# Function to map Gender into binary values
def map_gender_binary(gender):
    if gender == 'M':
        return 0  # Male
    elif gender == 'F':
        return 1  # Female
    else:
        return 2  # Unknown or Other

# Apply the function to create 'Gender_Binary' variable
X_train['Gender_Binary'] = X_train['Gender'].apply(map_gender_binary)
X_test['Gender_Binary'] = X_test['Gender'].apply(map_gender_binary)
X_val['Gender_Binary'] = X_val['Gender'].apply(map_gender_binary)
```

```python
def categorize_ime4_custom(count):
    if count == 0:
        return 'None'
    elif 1 <= count <= 2:
        return 'Few'
    elif 3 <= count <= 5:
        return 'Moderate'
    elif 6 <= count <= 8:
        return 'High'
    else:
        return 'Very High'

# Apply the function to create the new category
X_train['IME-4 Category'] = X_train['IME-4 Count'].apply(categorize_ime4_custom)
X_test['IME-4 Category'] = X_test['IME-4 Count'].apply(categorize_ime4_custom)
X_val['IME-4 Category'] = X_val['IME-4 Count'].apply(categorize_ime4_custom)

# 'Age Group': categorize into '<20', '20-40', '40-60', 60-80, '>80'
def age_group(age):
    if age < 18:
        return '<18'
    elif 19 <= age < 35:
        return '19-35'
    elif 36 <= age < 50:
        return '36-50'
    elif 51 <= age < 65:
        return '51-65'
    else:
        return '>66'

X_train['Age Group'] = X_train['Age at Injury'].apply(age_group)
X_test['Age Group'] = X_test['Age at Injury'].apply(age_group)
X_val['Age Group'] = X_val['Age at Injury'].apply(age_group)

# 'Wage Bracket': create categories for 'Average Weekly Wage'
def wage_bracket(wage):
    if wage < 500:
        return 'low'
    elif 500 <= wage < 1000:
        return 'moderate'
    elif 1000 <= wage < 1500:
        return 'high'
    elif 1500 <= wage < 2000:
        return 'very high'
    else:
        return 'exceptional'

X_train['Wage Bracket'] = X_train['Average Weekly Wage'].apply(wage_bracket)
X_test['Wage Bracket'] = X_test['Average Weekly Wage'].apply(wage_bracket)
X_val['Wage Bracket'] = X_val['Average Weekly Wage'].apply(wage_bracket)

# 'Recovery Time Group': categorize 'Days Since Accident' into intervals
def recovery_time_group(days):
    if pd.isna(days): 
        return 'Missing'
    if days < 30:
        return '<30 days'
    elif 30 <= days <= 90:
        return '30-90 days'
    else:
        return '>90 days'

X_train['Recovery Time Group'] = X_train['Days_Since_Accident'].apply(recovery_time_group)
X_test['Recovery Time Group'] = X_test['Days_Since_Accident'].apply(recovery_time_group)
X_val['Recovery Time Group'] = X_val['Days_Since_Accident'].apply(recovery_time_group)

# 'Processing Delay Severity': combine 'Days C-2 to C-3' + 'Days C-2 to Assembly' + 'Days C-3 to Assembly' into a single metric
X_train['Processing Delay'] = (X_train['Days C-2 to C-3'] + X_train['Days C-2 to Assembly'] + X_train['Days C-3 to Assembly']) / 3  # average of delays
X_test['Processing Delay'] = (X_test['Days C-2 to C-3'] + X_test['Days C-2 to Assembly'] + X_test['Days C-3 to Assembly']) / 3  # Average of delays
X_val['Processing Delay'] = (X_val['Days C-2 to C-3'] + X_val['Days C-2 to Assembly'] + X_val['Days C-3 to Assembly']) / 3  # Average of delays

def processing_delay_severity(delay):
    if pd.isna(delay):  
        return 'Missing'
    if delay < 2:
        return 'Low Delay'
    elif 2 <= delay <= 10:
        return 'Medium Delay'
    else:
        return 'High Delay'

X_train['Processing Delay Severity'] = X_train['Processing Delay'].apply(processing_delay_severity)
X_test['Processing Delay Severity'] = X_test['Processing Delay'].apply(processing_delay_severity)
X_val['Processing Delay Severity'] = X_val['Processing Delay'].apply(processing_delay_severity)

# Year Accident Grouping: Group accidents by year pre-2020 vs post-2020
def year_group(year):
    if pd.isna(year):  
        return 'Missing'
    if year < 2020:
        return 'Pre-2020'
    else:
        return 'Post-2020'

X_train['Year Accident Grouping'] = X_train['Accident_Year'].apply(year_group)
X_test['Year Accident Grouping'] = X_test['Accident_Year'].apply(year_group)
X_val['Year Accident Grouping'] = X_val['Accident_Year'].apply(year_group)

# check the first few rows to verify the newly created features
print(X_train[['Age Group', 'Wage Bracket', 'Recovery Time Group', 
           'Processing Delay Severity', 'Year Accident Grouping']].head())
```

- By combining accident years and combining delays into a single severity metric, we can identify more general dynamics like inefficiencies or outside factors like changes in regulations. These changes enhance the model's capacity to identify connections between attributes and results and make the data easier to analyse.

```python

from imblearn.over_sampling import ADASYN
from imblearn.under_sampling import RandomUnderSampler
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score, classification_report
from sklearn.preprocessing import LabelEncoder, RobustScaler
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import StratifiedGroupKFold, train_test_split


# Set random seed for reproducibility
np.random.seed(369)

# Reduce dataset size during cross-validation (retain 20% of data)
DATA_REDUCTION_PERCENTAGE = 0.3

# Identify numerical and categorical columns
numerical_columns = X_train.select_dtypes(include=np.number).columns.tolist()
categorical_columns = X_train.select_dtypes(exclude=np.number).columns.tolist()

# Define models
models = {
    'XGBoost': XGBClassifier(random_state=369, use_label_encoder=False, eval_metric='mlogloss')    
}

# Define hyperparameter grids with more parameters for tuning #are this the right parameters to optmize before define bagging
param_grids = {
    'XGBoost': {
        'n_estimators': [20, 30, 300],
        'max_depth': [3, 5],
        'learning_rate': [0.01, 0.1]
    }
    
}

# Instantiate preprocessing tools
label_encoders = {col: LabelEncoder() for col in categorical_columns}
scaler = RobustScaler()

# Cross-validation function

def cross_validate_with_group_stratified(model_name, model, param_grid, X_train, y_train, group_column):
    skf = StratifiedGroupKFold(n_splits=2, shuffle=True, random_state=369)
    best_model_result = {'best_f1_score': -1, 'best_classification_report': '', 'best_params': None, 'best_selected_features': {}}
    f1_scores = []
    best_model_per_fold = []

    for fold_idx, (train_idx, val_idx) in enumerate(skf.split(X_train, y_train, groups=X_train[group_column])):
        # Split into train and validation folds
        X_train_fold, X_val_fold = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_train_fold, y_val_fold = y_train.iloc[train_idx], y_train.iloc[val_idx]

        # Reduce dataset size with stratification
        X_train_fold, _, y_train_fold, _ = train_test_split(
            X_train_fold, y_train_fold, 
            train_size=DATA_REDUCTION_PERCENTAGE, 
            stratify=y_train_fold, 
            random_state=369
        )
        
        # Encode categorical features
        for col in categorical_columns:
            label_encoders[col].fit(X_train_fold[col])
            X_train_fold[col] = label_encoders[col].transform(X_train_fold[col])
            unseen_label = -1
            X_val_fold[col] = X_val_fold[col].apply(lambda x: x if x in label_encoders[col].classes_ else None)
            label_encoders[col].classes_ = np.append(label_encoders[col].classes_, None)
            X_val_fold[col] = X_val_fold[col].apply(lambda x: unseen_label if pd.isna(x) else label_encoders[col].transform([x])[0])

        # Scale numerical features
        scaler.fit(X_train_fold[numerical_columns])
        X_train_fold[numerical_columns] = scaler.transform(X_train_fold[numerical_columns])
        X_val_fold[numerical_columns] = scaler.transform(X_val_fold[numerical_columns])
        
        

        # Resample data using ADASYN
        adasyn = ADASYN(sampling_strategy='auto', random_state=369)
        X_train_fold, y_train_fold = adasyn.fit_resample(X_train_fold, y_train_fold)

        # Undersample majority classes
        undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=369)
        X_train_fold, y_train_fold = undersampler.fit_resample(X_train_fold, y_train_fold)
                   
        # Fit Random Forest for feature selection
        rf_selector = RandomForestClassifier(n_estimators=100, random_state=369, class_weight='balanced')
        rf_selector.fit(X_train_fold[numerical_columns], y_train_fold)
        
        # Get feature importances
        feature_importances = rf_selector.feature_importances_
        
        # Dynamically set threshold as the mean of feature importances
        mean_importance = np.mean(feature_importances)
        selected_numerical_columns = [
            numerical_columns[i] for i, importance in enumerate(feature_importances) if importance >= mean_importance
        ]
        
        #Chi-Square feature selection
        chi2_selector = SelectKBest(chi2, k='all')
        chi2_selector.fit(X_train_fold[categorical_columns], y_train_fold)
        
        # Get p-values for all categorical features
        p_values = chi2_selector.pvalues_
        
        # Apply the 10% p-value threshold
        threshold = 0.01
        selected_categorical_columns = [
            categorical_columns[i] for i, p_value in enumerate(p_values) if p_value <= threshold
        ]

        # Retain only selected features
        X_train_fold = pd.concat([X_train_fold[selected_numerical_columns], X_train_fold[selected_categorical_columns]], axis=1)
        X_val_fold = pd.concat([X_val_fold[selected_numerical_columns], X_val_fold[selected_categorical_columns]], axis=1)
        
        # Store selected features
        best_model_result['best_selected_features'][f'Fold {fold_idx+1}'] = {
            'numerical_features': selected_numerical_columns,
            'categorical_features': selected_categorical_columns
        }

        # Hyperparameter tuning
        grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring='f1_macro', cv=2, verbose=1)
        grid_search.fit(X_train_fold, y_train_fold)
        best_model = grid_search.best_estimator_
        best_model_per_fold.append(best_model)
        
        
        y_val_pred = best_model.predict(X_val_fold)
        f1 = f1_score(y_val_fold, y_val_pred, average='macro')
        # If the current model's F1 score is better than the best, update
        if f1 > best_model_result['best_f1_score']:
            best_model_result['best_f1_score'] = f1
            best_model_result['best_classification_report'] = classification_report(y_val_fold, y_val_pred)
            best_model_result['best_params'] = grid_search.best_params_

    return best_model_result
# Run cross-validation for each model
results = {}

for model_name, model in models.items():
    print(f"\nModel: {model_name}")
    best_result = cross_validate_with_group_stratified(
        model_name, model, param_grids[model_name], X_train, y_train, group_column='District Name'
    )
    # Print the best result for each model
    print(f"Best F1 Macro Score: {best_result['best_f1_score']:.4f}")
    print(f"Best Hyperparameters: {best_result['best_params']}")
    print("Classification Report:\n", best_result['best_classification_report'])
    
    # Print selected features
    for fold, features in best_result['best_selected_features'].items():
        print(f"{fold} - Selected Numerical Features: {features['numerical_features']}")
        print(f"{fold} - Selected Categorical Features: {features['categorical_features']}")
```
